```python
def quarter_find(x):
    if (x[6:10]=='2012'):
        if (x[3:5]=='07') | (x[3:5]=='08') | (x[3:5]=='09'):
            return x[3:5]+"-"+x[6:10]
```

```python
walmart['date_quarter_considered']=walmart.Date.apply(quarter_find)
```

```python
def super_bowl_decoded(x):
    if (x[6:10]=='2010') | (x[6:10]=='2011') | (x[6:10]=='2012') | (x[6:10]=='2013'):
        if (x[3:5]=='02'):
            if (x[0:2]=='12') | (x[0:2]=='11') | (x[0:2]=='10') | (x[0:2]=='08'):
                return 'Super Bowl'
```

```python
def labour_day(m):
        if (m[6:10]=='2010') | (m[6:10]=='2011') | (m[6:10]=='2012') | (m[6:10]=='2013'):
            if (m[3:5]=='09'):
                if (m[0:2]=='06') | (m[0:2]=='07') | (m[0:2]=='09') | (m[0:2]=='10'):
                    return 'Labour Day'
```

```python
def thanks_giving(x):
    if (x[3:5]=='11'):
        if (x[6:10]=='2010') | (x[6:10]=='2011') | (x[6:10]=='2012') | (x[6:10]=='2013'):
            if (x[0:2]=='26') | (x[0:2]=='25') | (x[0:2]=='23') | (x[0:2]=='29'):
                return 'Thanksgiving'
```

```python
def christmas(x):
    if (x[6:10]=='2010') | (x[6:10]=='2011') | (x[6:10]=='2012') | (x[6:10]=='2013'):
        if (x[3:5]=='12'):
            if (x[0:2]=='31') | (x[0:2]=='30') | (x[0:2]=='28') | (x[0:2]=='27'):
                return 'Christmas'
```

```python
walmart['Date_christmas']=walmart.Date.apply(christmas)
```

```python
walmart_christmas=walmart[walmart['Date_christmas']=='Christmas']
```
